```python
import pandas as pd
import os
```

```python
df = pd.read_csv('labels.csv')
```

```python
df.head(20)
```

```python
df.iloc[57,1] = 'bha'
```

```python
df_numerals = df.iloc[2:12, :]
print(df_numerals)
numerals_num = df_numerals.iloc[:,0].to_list()
numerals_label = df_numerals.iloc[:,1].to_list()
numeral_dict = dict(zip(numerals_num, numerals_label))
```

```python
df_vowels = df.iloc[17:29, :]
print(df_vowels)
vowel_num = df_vowels.iloc[:, 0].to_list()
print(vowel_num)
vowel_label = df_vowels.iloc[:, 1].to_list()
print(vowel_label)

vowel_dict = dict(zip(vowel_num, vowel_label))
print(vowel_dict)
```

```python
df_consonants = df.iloc[34:, :]
print(df_consonants)

consonant_num = df_consonants.iloc[:, 0].to_list()
consonant_label = df_consonants.iloc[:, 1].to_list()

consonant_dict = dict(zip(consonant_num, consonant_label))
print(consonant_dict)
```

```python
current_dir = os.getcwd()
new_dir = os.path.join(current_dir, r'devnagari')
os.makedirs(new_dir)
```

```python
def rename_folders(folder, name_dict):
    old_path = os.path.join(current_dir, r'nhcd\nhcd\{0}'.format(folder))
    print(old_path)
    for old_name, new_name in name_dict.items():
        old = os.path.join(old_path, str(old_name))
        print(old)
        new = os.path.join(new_dir, str(new_name))
        print(new)
        os.rename(old, new)
```

```python
rename_folders('numerals', numeral_dict)
rename_folders('vowels', vowel_dict)
rename_folders('consonants', consonant_dict)
```
